```python
from openpyxl import load_workbook
import pandas as pd

wb = load_workbook(filename='data/Stefan surgical 2019.xlsx', read_only=True)
```

```python
def load_demographics(wb):
  ws = wb['Demographics']
  data = ws.values
  next(data)
  cols = next(data)
  cols = [x.lower() for x in cols]
  data = list(data)
  df = pd.DataFrame(data, columns=cols)
  df = df.drop(['first name', 'second name'], axis=1)
  df = df.replace(r'\s*(.*?)\s*', r'\1', regex=True)
  df = df.set_index(['record number'])
  binary_cols = df.columns[5:]
  df[binary_cols] = df[binary_cols].replace(['Yes', 'No'], [1, 0])
  df[['covid status']] = df[['covid status']].replace(['Negative', 'Positive', 'N/A'], [0,1, None])
  return df
```

```python
def load_imaging(wb):
  ws = wb['Imaging']
  data = ws.values
  next(data)
  cols = next(data)
  cols = [x.lower() for x in cols]
  data = list(data)
  df = pd.DataFrame(data, columns=cols)
  binary_cols = df.columns[:-1]
  df[binary_cols] = df[binary_cols].replace(r'\s*(.*?)\s*', r'\1', regex=True)
  df[binary_cols] = df[binary_cols].replace(['Yes', 'No', 'NA'], [1, 0, None])
  df = df.set_index(['record number'])
  return df
```

```python
def load_management(wb):
  ws = wb['Management']
  data = ws.values
  next(data)
  cols = next(data)
  cols = [x.lower() for x in cols]
  cols[-1] = 'reoperated'
  data = list(data)
  df = pd.DataFrame(data, columns=cols)
  binary_cols = df.columns[1:]
  df[binary_cols] = df[binary_cols].replace(r'\s*(.*?)\s*', r'\1', regex=True)
  df[binary_cols] = df[binary_cols].replace(['Yes', 'No'], [1, 0])
  df = df.set_index(['record number'])
  return df
```

```python
def load_operations(wb):
  ws = wb['Operative data']
  data = ws.values
  next(data)
  cols = next(data)
  cols = [x.lower() for x in cols]
  cols[-1] = 'reoperated'
  data = list(data)
  df = pd.DataFrame(data, columns=cols)
  binary_cols = df.columns[1:]
  df[binary_cols] = df[binary_cols].replace(r'\s*(.*?)\s*', r'\1', regex=True)
  df[binary_cols] = df[binary_cols].replace(['Yes', 'No'], [1, 0])
  df = df.set_index(['record number'])
  return df
```
